### Change these to change the type of plot - int_map outputs an intensity map at the selected wavelength, wv_map outputs a map showing the brightest wavelength at each pixel

```python
int_map = 1
## for an intensity map at selected wavelength ##
wv_map = 1
## for a wavelength map at maximum intensity ##

if int_map == 1:
    wvlg_asked = input('Provide wavelength to plot at (in nm)')
```

```python
for filename in spectra_files:

    ## find code for each spectra to assign index in dataframe ##
    filecode = str(filename.split('_X')[0])
    index = data_info.index[data_info['code'] == filecode][0]

    ## for each x-y point, find wavelength of max intensity and add to data info dataframe ##
    if wv_map == 1:
        spectra = pd.read_csv(folder + '/' + filename, sep = '\t', names = ['Wavelength', 'Intensity'])
        max_int = max(spectra['Intensity'])
        max_wvl = spectra.iloc[spectra.index[spectra['Intensity'] == max_int][0]]['Wavelength']
        data_info.loc[index, 'Max Wavelength'] = max_wvl


    ## for each x-y point, find the intensity closest to selected wavelength and add to the data info dataframe ##
    if int_map == 1:
        spectra = pd.read_csv(folder + '/' + filename, sep = '\t', names = ['Wavelength', 'Intensity'])
        wvlg = min(spectra['Wavelength'], key=lambda x:abs(x-int(wvlg_asked)))
        indy = spectra.index[spectra['Wavelength'] == wvlg][0]
        intensity = spectra.iloc[indy]['Intensity']
        data_info.loc[index, 'Intensity'] = intensity

## create a 2D array containing intensity values in the x-y plane ##
int_array = []
wv_array = []
for yval in data_info['Y'].unique():
    index_list = data_info.index[data_info['Y'] == yval].tolist()
    if int_map == 1:
        int_array.append([data_info.iloc[index]['Intensity'] for index in index_list])
    if wv_map == 1:
        wv_array.append([data_info.iloc[index]['Max Wavelength'] for index in index_list])

## define X,Y,Z arrays for contour plot ##
x = list(data_info['X'].unique())
y = list(data_info['Y'].unique())

if int_map == 1:
    Z = int_array
    fig_int = plt.imshow(Z, cmap = 'inferno', extent = [min(x) - 0.5, max(x) + 0.5, min(y) - 0.5, max(y) + 0.5])
    plt.colorbar(label = 'Intensity (counts)')
    plt.xlabel(u' X (\u03bcm)', fontsize = 15)
    plt.ylabel(u' Y (\u03bcm)', fontsize = 15)
    plt.title(f'Intensity at {wvlg} nm')
    plt.show()
    plt.close()
    
    
if wv_map == 1:
    Z = wv_array
    fig_wv = plt.imshow(Z, cmap = turbo_cmap, extent = [min(x) - 0.5, max(x) + 0.5, min(y) - 0.5, max(y) + 0.5])
    plt.colorbar(label = 'Intensity (counts)')
    plt.xlabel(u' X (\u03bcm)', fontsize = 15)
    plt.ylabel(u' Y (\u03bcm)', fontsize = 15)
    plt.title('Wavelength of peak PL')
    plt.show()
    plt.close()



    
```
